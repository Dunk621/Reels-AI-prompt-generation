```python
# Choose a URL to embed in the QR code
qr_url = "https://www.emorygimg.com/"

# Print the chosen URL
print(f"The chosen URL for the QR code is: {qr_url}")
```

```python
import qrcode

# Create a QR code object
qr = qrcode.QRCode(
    version=1,
    error_correction=qrcode.constants.ERROR_CORRECT_L,
    box_size=10,
    border=4,
)

# Add the URL to the QR code
qr.add_data(qr_url)
qr.make(fit=True)
```

```python
# Create an image from the QR code object
img = qr.make_image(fill_color="black", back_color="white")

# Save the image to a file
img.save("python_qr_code.png")
```
